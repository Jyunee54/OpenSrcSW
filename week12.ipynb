```python
#실습1

import csv
f = open('file/seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
header = next(data)
max_temp = -999
max_date = ''
min_temp = 999
min_date = ''
range_temp =-999
range_date = ''

for row in data:
    if row[-1] == '':
        row[-1] = -999
    if row[-2] == '':
        row[-2] = 999
        
    row[-1] = float(row[-1])
    row[-2] = float(row[-2])
    
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
    if min_temp > row[-2]:
        min_date = row[0]
        min_temp = row[-2]
    if range_temp < row[-1]-row[-2]:
        range_date = row[0]
        range_temp = row[-1]-row[-2]

f.close()
print('일교차가 가장 큰 날')
print('날짜 : ', range_date)
print('최저기온이 가장 낮은 날')
print('날짜 : ', min_date)
print('최고기온이 가장 높은 날')
print('날짜 : ',max_date)
```
